#### 12. Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.

```python
ganhoPh = float(input("Ganho por hora?"))
horasM = float(input("Horas trabalhadas no mês?"))
```

#####  Calcule o salário bruto (horas * salario por hora)

```python
sBruto = ganhoPh * horasM
print(sBruto)
```

##### Calcule o desconto do IR (11% do salário bruto)

```python
descontoIR = 0.11
sDescontado = sBruto - (sBruto * descontoIR)
print(sDescontado)
```

##### Calcule o desconto do INSS (8% do salário bruto)

```python
descontoINSS = 0.08
sDescontado = sBruto - (sBruto * descontoINSS)
print(sDescontado)
```

##### Calcule o desconto do sindicato (5% do salário bruto)

```python
descontoSind = 0.05
sDescontado = sBruto - (sBruto * descontoSind)
print(sDescontado)
```

##### Calcule o salário líquido (salário bruto - descontos)

```python
descontos = sBruto * (descontoIR + descontoINSS + descontoSind)
sDescontado = sBruto - descontos
print(sDescontado)
```
